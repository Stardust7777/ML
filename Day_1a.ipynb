```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# Load the Iris dataset using pandas
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv(url, header=None, names=col_names)

# Convert to NumPy array
iris_np = iris.iloc[:, :-1].values  # Taking only the numerical columns

# Apply NumPy functions
mean_sepal_length = np.mean(iris_np[:, 0])
median_petal_width = np.median(iris_np[:, 3])
std_sepal_width = np.std(iris_np[:, 1])
max_petal_length = np.max(iris_np[:, 2])
min_sepal_width = np.min(iris_np[:, 1])
argmax_sepal_length = np.argmax(iris_np[:, 0])
unique_classes = np.unique(iris['class'])

# Reshaping the array (for demonstration)
reshaped_array = iris_np.reshape(-1, 2)  # Reshape it to two columns

# Display the calculated results
print(f"Mean of Sepal Length: {mean_sepal_length}")
print(f"Median of Petal Width: {median_petal_width}")
print(f"Standard Deviation of Sepal Width: {std_sepal_width}")
print(f"Maximum of Petal Length: {max_petal_length}")
print(f"Minimum of Sepal Width: {min_sepal_width}")
print(f"Index of Maximum Sepal Length: {argmax_sepal_length}")
print(f"Unique Classes in Dataset: {unique_classes}")
print(f"Reshaped Array: \n{reshaped_array}")

```

```python
#petal_grading

from ipywidgets import widgets, interactive
from matplotlib.widgets import CheckButtons

# Prepare color mapping for different Iris species
species = iris['class'].unique()
colors = ['blue', 'green', 'red']
species_colors = {species[i]: colors[i] for i in range(len(species))}

# Plotting with color-coded classes
plt.figure(figsize=(10, 7))

fig, ax = plt.subplots()


for spec in species:
    species_data = iris[iris['class'] == spec]
    # Scatter plot for Sepal Length vs Sepal Width
    if spec == 'Iris-setosa':
      s0 = ax.scatter(species_data['petal_length'], species_data['petal_width'], color=species_colors[spec], label=spec, alpha=0.7 )
    elif spec == 'Iris-versicolor':
      s1 = ax.scatter(species_data['petal_length'], species_data['petal_width'], color=species_colors[spec], label=spec, alpha=0.7)
    elif spec == 'Iris-virginica':
      s2 = ax.scatter(species_data['petal_length'], species_data['petal_width'], color=species_colors[spec], label=spec, alpha=0.7)

    # Linear trend line (regression)
    X = species_data[['petal_length']]
    y = species_data['petal_width']
    reg = LinearRegression().fit(X, y)
    plt.plot(X, reg.predict(X), color=species_colors[spec], linestyle='--')

lines_by_label = {s.get_label(): s for s in [s0, s1, s2]}
line_colors = [s.get_facecolor().flatten() for s in lines_by_label.values()]


rax = ax.inset_axes([0.0, 0.4, 0.3, 0.2])
check = CheckButtons(
    ax=rax,
    labels=lines_by_label.keys(),
    actives=[s.get_visible() for s in lines_by_label.values()],
    label_props={'color': line_colors},
    frame_props={'edgecolor': line_colors},
    check_props={'facecolor': line_colors},
)

def select_species(label):
    for line in lines_by_label.values():
        line.set_visible(line.get_label() == label)
    fig.canvas.draw_idle()

check.on_clicked(select_species)


# Adding labels and a title
plt.title("Petal Length vs Petal Width by Iris Species", fontsize=16)
plt.xlabel("Petal Length (cm)", fontsize=14)
plt.ylabel("Petal Width (cm)", fontsize=14)

# Grid and legend
plt.grid(True)
plt.legend(title="Iris Species", title_fontsize=12, fontsize=10)

# Enable interactivity: Hover over points to show coordinates
cursor = mplcursors.cursor(hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(
    f'Petal Length: {sel.target[0]:.2f}\nPetal Width: {sel.target[1]:.2f}'))

# Display the plot
plt.show()
```
